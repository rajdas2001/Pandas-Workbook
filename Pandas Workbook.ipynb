# Filtering Data

```python
df.loc[(df['Type 1']=='Fire') & (df['Type 2']=='Flying')]
```

```python
df.loc[(df['Type 1']=='Grass') | (df['Type 2']=='Poison')]
```

```python
df.loc[(df['Type 1']=='Grass') & (df['Type 2']=='Poison') & (df['HP']>70)]
```

```python
new_df=df.loc[(df['Type 1']=='Fire') & (df['Type 2']=='Flying')]
new_df=new_df.reset_index()
new_df
```

```python
new_df=df.loc[(df['Type 1']=='Fire') & (df['Type 2']=='Flying')]
new_df=new_df.reset_index()
new_df=new_df.drop(columns=['index','#'])
new_df
```

```python
df.loc[df['Name'].str.contains('Mega')]
```

```python
df.loc[~df['Name'].str.contains('Mega')] #for not containing the word 'mega' we use the tilde symbol (~)
```

# Regex Expressions

```python
import re
df.loc[df['Type 1'].str.contains('Fire|Grass', regex=True)]
```

```python
df.loc[df['Type 1'].str.contains('fire|grass', regex=True, flags=re.I)] #Includes Lower Case - fire and grass
```

```python
df.loc[df['Name'].str.contains('pi[a-z]*', regex=True, flags=re.I)]
```

```python
df.loc[df['Name'].str.contains('^pi[a-z]*', regex=True, flags=re.I)]
```

# Conditional Changes

```python
df.loc[df['Total']>500,['Generation','Legendary']]='Test Value'
df
```

```python
df.loc[df['Total']>500,['Legendary','Generation']]=['Test Value 1','Test Value 2']
df
```

# Aggregate Statistics (Groupby)

```python
df
```

```python
df.groupby(['Type 1']).mean().sort_values('Defense',ascending=False)
```

```python
df.groupby(['Type 1']).mean().sort_values('Attack',ascending=False)
```

```python
df.groupby(['Type 1']).sum()
```

```python
df['count']=1
df.groupby(['Type 1']).count()['count']
```

```python
df['count']=1
df.groupby(['Type 1','Type 2']).count()['count']
```
